## `follow_link` 方法

robobrowser 的 `follow_link` 方法可以点击链接并自动完成跳转

```python
import re
from robobrowser import RoboBrowser
```

```python
url = 'http://www.qq.com'
browser = RoboBrowser(history=True)
browser.open(url)
```

获取今日话题的 link

```python
today_top = browser.find(id='todaytop').a
print(today_top['href'])
```

进行 `follow_link` 跳转

```python
browser.follow_link(today_top)
```

打印标题

```python
title = browser.select('h2')[0]
print('*'*42)
print(title.text)
print('*'*42)
```

打印正文内容

```python
content = browser.find(id='articleContent').text
print(content)
```

## 注意

一般来说都是用 `find/select/find_all` 方法过滤出相应的链接，然后调用 `follow_link(object)` 的方式去点击该链接。
