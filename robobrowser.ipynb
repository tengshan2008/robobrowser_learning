```python
import re
from robobrowser import RoboBrowser
```

### open

打开一个网页

```python
browser = RoboBrowser(history=True)
browser.open('http://www.jms58.com/')
```

### get_link

获取特定文本的第一个链接

```python
magic = browser.get_link("玄幻奇幻")
magic_url = magic.get('href')
print(magic_url)
```

### get_links

获取特定文本的所有链接

```python
indexs = browser.get_links("目录")
for index in indexs:
    index_url = index.get('href')
    print(index_url)
```

### back

返回上n页

```python
# open magic page
browser.open(magic_url)
print(browser.url)
# back main page
browser.back(n=1)
print(browser.url)
```

### forward

前进n页

```python
browser.forward()
print(browser.url)
```

### find

获取元素，同 `BeautifulSoup::find`

```python
browser.find('div')
```

### find_all

获取元素，同 `BeautifulSoup::find_all`

```python
browser.find_all('div')
```

### get_form get_forms

获取表单

### submit_form

提交表单

```python
# <form name="articlesearch" method="post" action="http***search"></form>
form = browser.get_form(action="http://www.jms58.com/novel.php?action=search")
# <input type="text" name="searchkey" id="searchkey" class="text1" value="请输入***" onfocus="***" />
form['searchkey'].value = '超级兵王'
browser.submit_form(form)
```

### follow_link

直接访问标签中的链接，效果同`open`

```python
links = browser.get_links('的')
print(links)
for link in links:
    print(link.text)
    browser.follow_link(link)
    print(browser.url)
```

### select

获取元素，同 `BeautifulSoup::select`

```python
browser.select('tr')
```
